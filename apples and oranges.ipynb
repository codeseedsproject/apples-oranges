```python
# data analysis libraries and other basic imports
import pandas as pd
import numpy as np
from __future__ import division
import datetime as dt

# libraries for getting data sets
import Quandl
from pandas_datareader.data import DataReader

# plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
oj = Quandl.get("CHRIS/ICE_OJ1")
```

```python
oj_min = oj.ix[0].name
```

```python
aapl = DataReader('AAPL', 'yahoo', oj_min)
```

```python
price_df = pd.concat((aapl['Adj Close'], oj['Settle']), axis=1)
```

```python
price_df.columns = ['AAPL', 'OJ']
```

```python
price_df.head()
```

```python
price_df.plot()
```

```python
start = dt.datetime.strptime('2006-01-01', '%Y-%m-%d')
```

```python
price_df1 = price_df.loc[price_df.index >= start]
price_df2 = price_df.loc[price_df.index< start]
```

```python
price_df1.plot()
```

```python
price_df2.plot()
```

```python
sns.pairplot(price_df1)
```

```python
sns.pairplot(price_df2.dropna())
```

```python
sns.pairplot(price_df[price_df.AAPL>2])
```
